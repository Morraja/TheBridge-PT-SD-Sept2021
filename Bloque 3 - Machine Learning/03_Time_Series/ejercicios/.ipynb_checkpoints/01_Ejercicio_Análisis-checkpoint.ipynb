```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Cargamos los datos

```python
df = pd.read_csv('data/df_furniture.csv')
sfecha = pd.to_datetime(df['Month'])
df['new_date'] = pd.date_range(sfecha.min(),sfecha.max(), freq='MS')
df = df.set_index(df['new_date'])
df = df.sort_index()
df.head()
```

```python
df.columns = ['price']
```

## 4. ¿Cómo es la serie temporal? ¿Tiene tendencia? ¿Aprecias alguna estacionalidad?

Se ve que tiene una clara tendencia al alza, a pesar de una pequeña bajada en als acciones en el 2007.

También apreciamos estacionalidad, al cumplirse cierto patrón a lo largo del tiempo

```python
fig, ax = plt.subplots(figsize=(10, 7), dpi=100)
plt.plot(df['price']);
```

## 5. ¿Podrías demostrar de dos maneras estadísticas que la serie no es stationary?

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['price'].values, autolag = 'AIC')
print('AD_Fuller Method:', result[1])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Modelo aditivo:
result_add = seasonal_decompose(df['price'], model='additive', extrapolate_trend='freq')
# Modelo multiplicativo
result_mul = seasonal_decompose(df['price'], model='multiplicative', extrapolate_trend='freq')
```

```python
plt.rcParams.update({'figure.figsize': (10,10)})
result_add.plot();
```

```python
result_mul.plot();
```

```python
df['year'] = df.index.year

for year in df['year'].unique():
    print(df[df['year']==year].describe())
```

## 6. ¿Conoces algún método para representar mejor la tendencia?

```python
result_mul.plot();
```

```python
from pandas.plotting import autocorrelation_plot
show_trend = df['price'].values / result_mul.seasonal / result_mul.resid
plt.plot(show_trend);
```

## 7. Estaría bien visualizar todos los componentes de la serie temporal por separado

```python
result_mul.plot();
```

```python
result_add.plot();
```

## 8. Demuestra mediante una gráfica de autocorrelación que la serie es estacional

```python
from pandas.plotting import autocorrelation_plot
show_trend = df['price'].values / result_mul.trend / result_mul.resid
autocorrelation_plot(show_trend.tolist());
```
