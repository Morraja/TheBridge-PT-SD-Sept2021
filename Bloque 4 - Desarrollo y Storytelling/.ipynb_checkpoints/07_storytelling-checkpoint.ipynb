# Storytelling
> *Como contar una historia con datos.*
___


## I. Introducción

El storytelling consiste en **trazar una historia que conecte con el público**. El storytelling es una metodología que nos permite conectar con nuestro público, cliente, lector… mediante la emoción, ya que captamos su atención y les incentivamos a la acción (este es el aspecto más poderoso de la narración). Cuanto más rápido capte la atención del usuario más se involucrará con tu contenido. Cuanto más interactúen con tu contenido, mayores serán las posibilidades de conversión. Hay que contar historias y conseguir la unión mediante vínculos afectivos para crear sentimiento de pertenencia.

## II. Partes del proceso

### 1. Entiende el contexto

Para entender el contexto lo que se necesita conocer es la audiencia a la que va dirigido, son perfiles más tecnicos o menos, son personas mas junior o más senior, quieren hacer algo o más bien conocer algo, son preguntas que tienes que contestarte para poder entender bien el contexto. Aquí podremos ver si necesitamos hacer algo exploratorio o explicativo por ejemplo.

Además tienes que pensar en los datos que vas a necesitar para apoyar tu historia, para ello tienes que entender de que datos puedes disponer o de si puedes añadir más información a ellos para poder ilustrar bien tu discurso.

Trabajar en papel en una pequeña versión de lo que se quiere hacer ayuda a ver el e2e.

**Who?**
- Cuanto más específico sea sobre quién es su público, mejor posición tendrá para lograr una comunicación exitosa. 
- La relación que tienes con la audiencia también importa y cómo esperas que te perciban. ¿Confían ya en ti como experto o necesitas trabajar para establecer tu credibilidad?

**What?**

- ¿Qué necesitas que sepa o haga tu audiencia? Si se limita a presentar datos, es fácil que la audiencia diga "Oh, eso es interesante" y pase a lo siguiente. Pero si pide una acción, su audiencia tiene que tomar la decisión de cumplirla o no. Cuando no sea apropiado recomendar explícitamente una acción, fomente la discusión hacia una.

**How?**
- Una vez que hayamos articulado quién es nuestro público y qué necesitamos que sepa o haga, podemos recurrir a los datos. Los datos se convierten en pruebas de apoyo de la historia que se va a construir y contar.

### 2. Escoge la visualización adecuada 

Mostrar los datos de manera diferente puede focalizar y facilitar la observación de los hechos que queremos resaltar.

[Python Graph Gallery](https://www.python-graph-gallery.com/)

### 3. Elimina lo innecesario (cluttering)

Elimina elementos que no incrementan la información que proporcionas. Dado que estos elementos a veces pueden complicar la visualización más de los necesario. Tendemos a querer meter en un gráfico la mayor información posible y esto hace que nos perdamos en lo que queremos realmente hacer notar o ver, por ello es recomendable eliminar todo aquello que no aporta información a lo que queremos mostrar.

Cada uno de los elementos que se añaden a la página o a la pantalla supone una carga cognitiva para el público. La carga cognitiva es el esfuerzo mental necesario para aprender una nueva información.

¿Qué es el **clutter**? Son elementos visuales que ocupan espacio pero no aumentan la comprensión.

Cuando nuestros elementos visuales o gráficas parecen complicados, corremos el riesgo de que nuestro público decida que no quiere dedicar tiempo a entender lo que le estamos mostrando, momento en el que habremos perdido nuestra capacidad de comunicación con él.

**Los principios de Gestallt** de la percepción visual **pueden ayudarte a entender cómo ve tu público**:

- **Proximity**: Tendemos a pensar que los objetos que están físicamente cerca forman parte de un grupo.
- **Similarity**: Los objectos de color, forma, tamaño u orientación similares se perciben como relacionados o pertenecientes a un grupo.
- **Enclosur**e: Pensamos que los objetos que están físicamente juntos forman parte de un grupo.
- **Closure**: El concepto de cierre dice que a las personas nos gusta que las cosas sean sencillas y encajen en las construcciones que ya tenemos en la cabeza; por ejemplo, cuando faltan partes de un conjunto, nuestros ojos rellenan los huecos.
- **Continuity**: El principio de continuidad es similar al de cierre: al mirar objetos, nuestros ojos buscan el camino más suave y crean de forma natural una continuidad en lo que vemos, incluso cuando no existe explícitamente.
- **Connection**: Tendemos a pensar que los objetos que están conectados físicamente forman parte de un grupo.

### 4. Focaliza la atención y cuenta una historia

Es importante tener claro el propósito de la historia y tratar de conectar con la audiencia para que se haga partícipe de tu historia tambien añadiendo historias personales o humor. Además de que hay que estructurar bien la historia, que tengan una introducción al problema, desarrollar la idea principal, mostrar resultados y exponer las conclusiones o decisiones además de finalmente añadir las referencias.

- Destacar cosas importantes.
- Eliminar distracciones.
- Comprender lo que el público quiere hacer con los datos (función) y crear una visualización (forma) que lo permita.
- No complicar en exceso, para que los diseños sean accesibles a todo el mundo.
- Utilizar el texto para etiquetar, presentar, explicar, reforzar, destacar, recomendar y, en definitiva, contar una historia.
- Hacer gráficos que sean estéticamente agradables.

## III. Ejemplo en código

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns;
sns.set_style()

%matplotlib inline
```

```python
df2 = pd.DataFrame({
    'Month':['January','February','March','April','May','June','July','August','September','October','November','December'],
    'Ticket_Volume_Received': [160,184,241,149,180,161,132,202,160,139,149,177],
    'Ticket_Volume_Processed':[160,184,237,148,181,150,123,156,126,104,124,140]
})
names = ['January','February','March','April','May','June','July','August','September','October','November','December']
names2 = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# Stacked-Bar
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

sns.barplot(x=df2.index,y=df2['Ticket_Volume_Received'],data=df2,label = 'Calls Logged',color='#4848FF',ax=ax)
sns.barplot(x=df2.index,y=df2['Ticket_Volume_Processed'],data=df2,label = 'Calls Resolved',color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=14,ha='right')

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),fancybox=True, shadow=False, ncol=5,fontsize=12)
plt.grid(axis='y')
plt.ylabel('')
plt.ylim(0,300)
plt.xticks(ticks=range(0, 12), labels=list(names));
```

```python
# Serie Temporal
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

sns.lineplot(x=df2.index,y=df2['Ticket_Volume_Received'],data=df2,marker='o',markersize=10,lw=4,label = 'Calls Logged',color='#4848FF',ax=ax)
sns.lineplot(x=df2.index,y=df2['Ticket_Volume_Processed'],data=df2,marker='s',markersize=10,lw=4,label = 'Calls Resolved',color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=14,ha='right')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),fancybox=True, shadow=False, ncol=5,fontsize=12)
plt.grid(axis='y')
plt.ylabel('')
plt.ylim(0,300);
plt.xticks(ticks=range(0, 12), labels=list(names));
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)
# Variables that define subsets of the data, which will be drawn on separate facets in the grid.
sns.lineplot(df2.index,y=df2['Ticket_Volume_Received'],marker='o',markersize=10,data=df2,lw=4,color='#4848FF',ax=ax)
sns.lineplot(df2.index,y=df2['Ticket_Volume_Processed'],marker='s',markersize=10,data=df2,lw=4,color='#FF6800',ax=ax)
plt.title('Monthly Volume of IT calls Logged and Resolved',fontsize=18,ha='right',color='#616161')
sns.despine()
# Change the colors of the left and bottom borders (fade into the background)
ax.spines['left'].set_color('lightgray')
ax.spines['bottom'].set_color('lightgray')
ax.text(12.6, 180, 'Logged', ha='right', va='top', fontsize=22,color='#4848FF')
ax.text(12.8, 145, 'Resolved', ha='right', va='top', fontsize=22,color='#FF6800')
plt.ylabel('')
plt.ylim(0,300)
plt.xticks(ticks=range(0, 12), labels=list(names2));
plt.ylabel('Volume',fontsize=16,color='#616161');
```
