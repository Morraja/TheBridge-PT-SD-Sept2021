# Gráficos de líneas simples

Quizás la más simple de todas las gráficas es la visualización de una sola función $y = f(x)$.

En este notebook, nos adentraremos en el mundo de las visualizaciones mediante la creación una simple gráfica de eeste tipo.
Al igual que en el resto de notebooks de este tema, comenzaremos estableciendo la configuración para representar las gráficas e importaremos los paquetes que usaremos:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

Para todas las visualizaciones de Matplotlib, comenzaremos creando una figura y sus ejes, cuya forma más básica seería algo así:

```python
fig = plt.figure()
ax = plt.axes()
```

En Matplotlib, la figura (una instancia de la clase `` plt.Figure ``) se puede entender como un contenedor único que contiene todos los objetos que representan ejes, gráficos, texto y etiquetas.

Los ejes (una instancia de la clase `` plt.Axes ``) es lo que vemos arriba: un cuadro delimitador con marcas y etiquetas, que eventualmente contendrá los elementos de la gráfica que componen nuestra visualización.

A lo largo del notebook, usaremos la variable ``fig`` para referirnos a una instancia de figura y ``ax`` para referirnos a una instancia de ejes o un grupo de instancias de ejes.

Una vez hayamos creado los ejes, podremos usar la función ``ax.plot`` para graficar algunos datos. Por ejemplo, comencemos por una senoidal simple:

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

Como vemos, comenzamos crando la figura y los ejes. Después nos creamos el vector de las X, que serán las entradas del sistema que queremos representar y, finalmente, pintamos sobre nuestro lienzo la salida de la función que queremos, que será el vector de elementos correspondientes a aplicar la función sobre nuestra entrada.

Alternativamente, podemos usar la otra interfaz que habíamos comentado en el notebook anterior, y dejar que la figura y los ejes se creen de forma transparente para nosotros:

```python
plt.plot(x, np.sin(x));
```

Si quisiéramos crear una única figura con más de una línea, podríamos simplemente llamar a la función ``plot`` más de una vez:

```python
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

Y ya lo tendríamos, así de simple es graficar funciones simples en Matplotlib.

Pero, normalmente, querremos tener un poco más de control sobre nuestros gráficos, por ello, pasemos ahora a ver con un poco más de detalle cómo controlar la apariencia de los ejes y las líneas.

## Tuneando gráficas: Colores y estilos de línea

Una de las primeras cosas que se nos viene a la mente a la hora de realizar un gráfico de más de una línea, es cómo diferenciar cada una de ellas, es decir, cómo elegir el color y el estilo de cada línea.

Esto lo podemos hacer a través de un parámetro más en nuestra función `` plt.plot () ``, que está pensada para que podamos modificar múltiples aspectos de nuestras gráficas, para lo que tendremos argumentos específicos.

En este caso, para ajustar el color utilizaremos el parámetro ``color``, el cual acepta un string o una tupla que represente virtualmente el color que queremos que lleve esa línea.

Veamos a continuación sobre un ejemplo, las diferentes formas que tenemos de llamar a un color:

```python
plt.plot(x, np.sin(x - 0), color='blue')        # por nombre (existen algunos definidos)
plt.plot(x, np.sin(x - 1), color='g')           # código corto de color (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Código hexadecimal (RRGGBB desde 00 a FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # tupla de RGB, con valores de 0 a 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # todos los colores soportados por HTML
```

Si no especificamos ningún color, Matplotlib lo hará por nosotros recorriendo automáticamente un conjunto de colores predeterminados para varias líneas.


Por otra parte, además del color, otra de las características de nuestras líneas que nos permitirá diferenciar cada representación, será el estilo de línea, el cual puede ser especificado mediante el parámetro ``linestyle``:

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# También podríamos utilizar atajos de nombre:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

Además, para facilitarnos la vida a los que somos un poco vagos (o eficientes, depende de cómo lo mires), Matplotlib nos permite combinar el estilo de línea y el color en un único parámetro sin nombre de la función ``plt.plot()``, del siguiente modo:

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

Estos códigos de color de un solo carácter reflejan las abreviaturas estándar en los sistemas de color RGB (rojo / verde / azul) (inglés: Red/Green/Blue) y CMYK (cian / magenta / amarillo / negro) (inglés: Cyan/Magenta/Yellow/blacK), que se utilizan comúnmente para los colores de los gráficos digitales.

Hay muchos otros argumentos de palabras clave que se pueden utilizar para ajustar la apariencia de la gráfica, que vienen recogidos en la documentación de la función ``plt.plot()``, la cual podemos acceder mediante código:

## Tuneando gráficas: Límites de los ejes

Matplotlib hace un trabajo decente al elegir los límites de ejes predeterminados para nuestras gráficas, pero a veces es bueno tener un control más preciso.
La forma más básica de ajustar los límites del eje es usar los métodos `` plt.xlim () `` y `` plt.ylim () ``:

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

Si por alguna razón quisiéramos que cualquiera de los ejes se mostrase al revés, simplemente podríamos invertir el orden de los argumentos:

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

Un método relacionado que nos puede resultar muy útil es `` plt.axis () `` (fíjate que este método es "axis" con ``i``, no lo confundas con el método "axes" con ``e``).

El método `` plt.axis () `` nos permite establecer los límites `` x `` e `` y `` con una sola llamada, pasándole como argumento una lista que especifica los límites de los ejes como `` [xmin, xmax, ymin, ymax] ``:

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

El método ``plt.axis()`` va más allá de esto, permitiéndonos hacer cosas como ajustar automáticamente los límites alrededor de la gráfica actual:

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

Permite incluso especificaciones de nivel superior, como asegurarnos mantener una relación de aspecto igual que la de nuestra pantalla, de modo que una unidad en `` x `` sea igual a una unidad en `` y ``:

```python
plt.plot(x, np.sin(x))
plt.axis("equal");
```

## Etiquetando las gráficas


Como última parte de este notebook, veremos brevemente el etiquetado en los gráficos: títulos, etiquetas de ejes y leyendas simples.

Los títulos y las etiquetas de los ejes son las etiquetas más simples de este tipo; existiendo métodos, y argumentos de otros, que se pueden usar para configurarlas rápidamente:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

La posición, el tamaño y el estilo de estas etiquetas se pueden ajustar utilizando argumentos opcionales para la función.

Para obtener más información, te recomiendo que consultes la documentación de Matplotlib y los docstrings de cada una de estas funciones.

Cuando se muestran varias líneas dentro de un solo eje, puede resultar útil crear una leyenda que etiquete cada tipo de línea.
Una vez más, Matplotlib tiene una forma integrada de crear rápidamente dicha leyenda.

Se hace a través del método `` plt.legend () ``. Aunque hay varias formas válidas de usar esto, una de las más sencillas sería especificar la etiqueta de cada línea usando la palabra clave `` label `` de la función de ``plt.plot()``:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();

plt.title("Sin and Cos")
plt.xlabel("X")
plt.ylabel("sin & cos")
```

Como puedes ver, la función ``plt.legend()`` realiza un seguimiento del color y estilo de la línea, y los combina con la etiqueta correcta.

Puedes obtener más información sobre cómo especificar y dar formato a las leyendas de los gráficos en el docstring de `` plt.legend ``. Además, cubriremos algunas opciones de leyenda más avanzadas más adelante.

## Interfaz MATLAB → Interfaz Orientada a Objetos 

Mientras que la mayoría de las funciones ``plt`` se traducen directamente a métodos ``ax`` (como `plt.plot ()` → `ax.plot ()`, `plt.legend () → ax.legend ()`, etc.), no es así para todos los comandos. En particular, las funciones para establecer límites, etiquetas y títulos sufren cierta modificación en su nombre.

Para realizar la transición entre funciones de estilo MATLAB y métodos orientados a objetos, tendremos que realizar los siguientes cambios:

- ``plt.xlabel()`` → ``ax.set_xlabel()``
- ``plt.ylabel()`` → ``ax.set_ylabel()``
- ``plt.xlim()`` → ``ax.set_xlim()``
- ``plt.ylim()`` → ``ax.set_ylim()``
- ``plt.title()`` → ``ax.set_title()``


En la interfaz orientada a objetos, en lugar de llamar a estas funciones individualmente, suele ser más conveniente usar el método ``ax.set ()`` para establecer todas estas propiedades a la vez:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10),
       ylim=(-2, 2),
       xlabel='x',
       ylabel='sin(x)',
       title='A Simple Plot');
```
